# QUESTION 10

** Hard: How many people have a credit card that expires in 2025? **

```python
## before we solve the challenge, we shall take a look at how the expiry date is defined in the "CC Expiry Date" Column below

ecomm.head(1)
```

```python
## WE WONT BE ABLE TO FILTER THE DATA TO DERIVE ACCURATELY THE YEAR OF THE EXPIRY DATE BECAUSE OF HOW THE EXPIRY DATE IS DEFINED
## SO WE SHALL WRITE A FUNCTION THAT WILL EXTRACT THE LAST 2 DIGITS OF THE EXPIRY DATE. FOR EXAMPLE "02/20" WILL BE EXTRACTED
## AND THE RESULT WILL BE "20" WHICH MEANS THE YEAR OF THE EXPIRY DATE IS 2020

## FUNCTION
def exp_date(x):
    x = list(x)      #CONVERT THE DATE TO A LIST AND STORE IN VARIABLE X
    x = x[3] + x[4]  #CONCATENATE INDEX 3 AND 4 OF THE LIST TO FORM A SINGLE STRING AND STORE IN VARIABLE X
    x = int(x)       #CONVERT THE VARIABLE INTO AN INTEGER AND STORE IN X
    return x
    
```

```python
# NEXT STEP, WE SHALL CREATE A NEW COLUMN CALLED "new_exp_date" ON OUR DATA FRAME AND APPLY THE LAMBDA FUNCTION ON "exp_date(x)"
# INSIDE THE NEW COLUMN WHICH IS TO BE CREATED

ecomm["new_exp_date"] = ecomm["CC Exp Date"].apply(lambda x: exp_date(x)) 
```

```python
# THE NEW COLUMN "new_exp_date" is the last column on the Data Frame 

ecomm.head(1)
```

```python
## TO FIND THE NUMBER OF PEOPLE WHOSE CREDIT WILL EXPIRE IN 2025, WE SHALL FILTER THE "new_exp_date" column on the Year "25".
## THEREAFTER WE SHALL APPLY THE COUNT "Sum()" FUNCTION

sum(ecomm["new_exp_date"] == 25)
```

```python
## There are 1033 people whose credit card have expiry date of 2025
```

# QUESTION 11

** Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
ecomm.head(1)
```

```python
# TO BEGIN THIS CHALLENGE, WE WILL HAVE TO CONVERT THE EMAIL ADDRESSES "AAA@yahoo.com" TO THIER HOST NAMES "yahoo.com"
# IN ORDER TO ACHIEVE THIS, WE SHALL WRITE A FUNCTION THAT WILL REMOVE THE EMAIL NAME AND RETURN ONLY THE HOST NAME "Yahoo.com"

def domain_name(x):
    x = x.split('@')      #Split the string when it gets to the word "@"
    x = x[1]              #return index 1 and store in variable x
    x = str(x)            #Convert the list to a string and store in x
    return x
```

```python
domain_name("seun@yahoo.com") # This is to test if our function returns the right result we intend for it to return
```

```python
# NEXT STEP, WE SHALL CREATE A NEW COLUMN CALLED "new_domain_name" ON OUR DATA FRAME AND APPLY THE LAMBDA FUNCTION ON 
# THE ACTUAL FUNCTION "domain_name(x)" INSIDE THE NEW COLUMN WHICH IS TO BE CREATED

ecomm["new_domain_name"] = ecomm["Email"].apply(lambda x: domain_name(x)) 
```

```python
# The NEW COLUMN CALLED "new_domain_name" IS THE LAST COLUMN ON THE DATA FRAM BELOW

ecomm.head(1)
```

```python
# TO SOLVE THE CHALLENGE, FIRSTLY, WE SHALL DETERMINE THE NUMBER OF TIMES THE HOST NAMES IN THE "new_domain_name" COLUMN OCCURED
# BY USING THE "value_counts()" FUNCTION

ecomm["new_domain_name"].value_counts()
```

```python
# NEXT, WE SHALL DETERMINE THE TOP FIVE(5) HOST NAMES BY USING THE "head()" function

ecomm["new_domain_name"].value_counts().head(5)
```
